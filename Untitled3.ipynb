```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Weather Data.csv')

# Replace zeros and null values with the median
for column in df.columns:
    median = df[column].median()
    df[column].replace(0, median, inplace=True)
    df[column].fillna(median, inplace=True)

# Define the feature matrix X and target variable y
X = df.drop('Avg Temperature', axis=1)  # Use all columns except 'Avg Temperature' as features
y = df['Avg Temperature']  # Use 'Avg Temperature' as the target variable

# Fit a Ridge regression model to find feature importances
ridge = Ridge(alpha=1.0)
ridge.fit(X, y)

# Get the feature importances
importances = np.abs(ridge.coef_)

# Create a DataFrame to hold the feature names and their importances
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': importances})

# Sort the features by importance
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

# Select the top 12 features based on importance
selected_features = feature_importances.head(12)['Feature']

# Subset the feature matrix to include only the selected features
X_selected = X[selected_features]

# Scale the features using MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_selected)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize and train a Bagging Regressor
bagging_regressor = BaggingRegressor(n_estimators=100, random_state=42)
bagging_regressor.fit(X_train, y_train)

# Predict on training and testing sets
y_train_pred = bagging_regressor.predict(X_train)
y_test_pred = bagging_regressor.predict(X_test)

# Calculate metrics for training and testing sets
train_r2 = r2_score(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
train_mae = mean_absolute_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)

test_r2 = r2_score(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)

# Perform cross-validation and calculate metrics
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_r2_scores = []
cv_mse_scores = []
cv_mae_scores = []
cv_rmse_scores = []

for train_index, test_index in kf.split(X_scaled):
    X_train_cv, X_test_cv = X_scaled[train_index], X_scaled[test_index]
    y_train_cv, y_test_cv = y[train_index], y[test_index]

    bagging_regressor.fit(X_train_cv, y_train_cv)
    y_pred_cv = bagging_regressor.predict(X_test_cv)

    cv_r2_scores.append(r2_score(y_test_cv, y_pred_cv))
    cv_mse_scores.append(mean_squared_error(y_test_cv, y_pred_cv))
    cv_mae_scores.append(mean_absolute_error(y_test_cv, y_pred_cv))
    cv_rmse_scores.append(np.sqrt(mean_squared_error(y_test_cv, y_pred_cv)))

# Output the results
print(f"Selected 12 features:\n{selected_features}\n")

print("Training set metrics:")
print(f"R²: {train_r2}")
print(f"MSE: {train_mse}")
print(f"MAE: {train_mae}")
print(f"RMSE: {train_rmse}\n")

print("Testing set metrics:")
print(f"R²: {test_r2}")
print(f"MSE: {test_mse}")
print(f"MAE: {test_mae}")
print(f"RMSE: {test_rmse}\n")

print("Cross-validation metrics:")
print(f"R² scores: {cv_r2_scores}")
print(f"Mean R² score: {np.mean(cv_r2_scores)}\n")
print(f"MSE scores: {cv_mse_scores}")
print(f"Mean MSE score: {np.mean(cv_mse_scores)}\n")
print(f"MAE scores: {cv_mae_scores}")
print(f"Mean MAE score: {np.mean(cv_mae_scores)}\n")
print(f"RMSE scores: {cv_rmse_scores}")
print(f"Mean RMSE score: {np.mean(cv_rmse_scores)}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Weather Data.csv')

# Replace zeros and null values with the median
for column in df.columns:
    median = df[column].median()
    df[column].replace(0, median, inplace=True)
    df[column].fillna(median, inplace=True)

# Define the feature matrix X and target variable y
X = df.drop('Avg Temperature', axis=1)  # Use all columns except 'Avg Temperature' as features
y = df['Avg Temperature']  # Use 'Avg Temperature' as the target variable

# Fit a Ridge regression model to find feature importances
ridge = Ridge(alpha=1.0)
ridge.fit(X, y)

# Get the feature importances
importances = np.abs(ridge.coef_)

# Create a DataFrame to hold the feature names and their importances
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': importances})

# Sort the features by importance
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

# Select the top 12 features based on importance
selected_features = feature_importances.head(12)['Feature']

# Subset the feature matrix to include only the selected features
X_selected = X[selected_features]

# Scale the features using MinMaxScaler
scaler_all = MinMaxScaler()
X_scaled_all = scaler_all.fit_transform(X)

scaler_selected = MinMaxScaler()
X_scaled_selected = scaler_selected.fit_transform(X_selected)

# Define a function to evaluate the model
def evaluate_model(X_scaled, y, description):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    # Initialize and train a Bagging Regressor
    bagging_regressor = BaggingRegressor(n_estimators=100, random_state=42)
    bagging_regressor.fit(X_train, y_train)

    # Predict on training and testing sets
    y_train_pred = bagging_regressor.predict(X_train)
    y_test_pred = bagging_regressor.predict(X_test)

    # Calculate metrics for training and testing sets
    train_r2 = r2_score(y_train, y_train_pred)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_rmse = np.sqrt(train_mse)

    test_r2 = r2_score(y_test, y_test_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_rmse = np.sqrt(test_mse)

    # Perform cross-validation and calculate metrics
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_r2_scores = []
    cv_mse_scores = []
    cv_mae_scores = []
    cv_rmse_scores = []

    for train_index, test_index in kf.split(X_scaled):
        X_train_cv, X_test_cv = X_scaled[train_index], X_scaled[test_index]
        y_train_cv, y_test_cv = y[train_index], y[test_index]

        bagging_regressor.fit(X_train_cv, y_train_cv)
        y_pred_cv = bagging_regressor.predict(X_test_cv)

        cv_r2_scores.append(r2_score(y_test_cv, y_pred_cv))
        cv_mse_scores.append(mean_squared_error(y_test_cv, y_pred_cv))
        cv_mae_scores.append(mean_absolute_error(y_test_cv, y_pred_cv))
        cv_rmse_scores.append(np.sqrt(mean_squared_error(y_test_cv, y_pred_cv)))

    # Output the results
    print(f"Results for {description}:")
    print(f"Selected 12 features:\n{selected_features}\n")

    print("Training set metrics:")
    print(f"R²: {train_r2}")
    print(f"MSE: {train_mse}")
    print(f"MAE: {train_mae}")
    print(f"RMSE: {train_rmse}\n")

    print("Testing set metrics:")
    print(f"R²: {test_r2}")
    print(f"MSE: {test_mse}")
    print(f"MAE: {test_mae}")
    print(f"RMSE: {test_rmse}\n")

    print("Cross-validation metrics:")
    print(f"R² scores: {cv_r2_scores}")
    print(f"Mean R² score: {np.mean(cv_r2_scores)}\n")
    print(f"MSE scores: {cv_mse_scores}")
    print(f"Mean MSE score: {np.mean(cv_mse_scores)}\n")
    print(f"MAE scores: {cv_mae_scores}")
    print(f"Mean MAE score: {np.mean(cv_mae_scores)}\n")
    print(f"RMSE scores: {cv_rmse_scores}")
    print(f"Mean RMSE score: {np.mean(cv_rmse_scores)}")
    print("\n" + "-"*50 + "\n")

# Evaluate the model with all features
evaluate_model(X_scaled_all, y, "All Features")

# Evaluate the model with selected features
evaluate_model(X_scaled_selected, y, "Selected Features")
```
